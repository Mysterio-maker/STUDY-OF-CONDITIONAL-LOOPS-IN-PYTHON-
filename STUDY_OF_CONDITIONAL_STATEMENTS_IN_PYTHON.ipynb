```python
# Python program for date validation and increment

date = input("Enter the date in format {dd/mm/yyyy}: ")

try:
    dd, mm, yy = date.split("/")
    dd = int(dd)
    mm = int(mm)
    yy = int(yy)

    if mm < 1 or mm > 12:
        print("Date is invalid.")

    else:
        # Days in month
        if mm in (1, 3, 5, 7, 8, 10, 12):
            max_days = 31
        elif mm in (4, 6, 9, 11):
            max_days = 30
        else:  # February
            if (yy % 4 == 0 and yy % 100 != 0) or (yy % 400 == 0):
                max_days = 29
            else:
                max_days = 28

        if dd < 1 or dd > max_days:
            print("Date is invalid.")
        else:
            # Increment date
            if dd == max_days:
                dd = 1
                if mm == 12:
                    mm = 1
                    yy += 1
                else:
                    mm += 1
            else:
                dd += 1

            print(f"The incremented date is: {dd:02d}/{mm:02d}/{yy}")

except:
    print("Invalid format! Please enter date in dd/mm/yyyy format.")
```

```python
# Leap year
year = int(input("Enter year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap Year")
else:
    print("Not Leap Year")
```

```python
# Salary Calculation with Allowances

basic = int(input("Enter the Basic Salary: "))

if basic <= 10000:
    da = basic * 20 / 100
    hra = basic * 80 / 100
elif basic <= 20000:
    da = basic * 25 / 100
    hra = basic * 90 / 100
else:
    da = basic * 30 / 100
    hra = basic * 95 / 100

gross = basic + da + hra

print("Gross Salary:", gross)
```

```python
# Income Tax Calculation Program

salary = int(input("Enter your annual income: "))

if salary <= 250000:
    tax = 0

elif salary <= 500000:
    tax = (salary - 250000) * 0.05

elif salary <= 1000000:
    tax = (250000 * 0.05) + (salary - 500000) * 0.20

else:
    tax = (250000 * 0.05) + (500000 * 0.20) + (salary - 1000000) * 0.30

print("Total Tax to be Paid:", tax)
```
